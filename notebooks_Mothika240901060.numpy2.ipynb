```python
import numpy as np
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
print("Original array:\n",arr)
```

```python
#Returns every other rows
print("\nEvery other rows:\n",arr[0:3:2])
```

```python
#return every other element from the entire array
arr=np.array([1,2,3,4,5,6,7])
print("\nOriginal array:",arr)
print("\nReturns every other array[::2]\n",arr[::2])
```

```python
#slicing array
arr=np.array([[-1,2,0,4],[4,-0.5,6,0],[2.6,0,7,8],[3,-7,4,2.0]])
temp=arr[:2,:3]
print("\nArray with first 2 rows and 3 columns:\n",temp)
```

```python
#integer array indexing
tem=arr[[0,1,2,3],[3,2,1,0]]
print("\nElement at indices (0,3),(1,2),(2,1),(3,0)," "(3,0):\n",tem)
```

```python
#boolean array indices
cond=arr>2
temp=arr[cond]
print("\nElements greater than 2:\n",temp)
```

```python
import numpy as np
#joining two array
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate([arr1,arr2])
print(arr)
```

```python
#Horizontal join
arr3=np.hstack((arr1,arr2))
print(arr3)
#Vertical join
arr4=np.vstack((arr1,arr2))
print(arr4)
#Depth join
arr5=np.dstack((arr1,arr2))
print(arr5)
```

```python
#splitting of array
arr=np.array([1,2,3,4,5,6])
arr6=np.array_split(arr,3)
print(arr)
print(arr6)
print(arr6[0])
print(arr6[1])
print(arr6[2])
```
